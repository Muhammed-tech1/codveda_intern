#### import the required libraries

```python
# Data manipulation libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data loading

```python
data = pd.read_csv('house_data.csv')
```

```python
# print the first five rows of the data
data.head()
```

```python
print('This is the info of the data:\n', data.info())
```

### Compute summary statistics (mean, median, variance, etc).

```python
print('This the summary of the data:\n', data.describe(),'\n')
```

```python
# get the mean of the data
print('This the mean value of the dataset:\n', data.mean(), '\n')
```

```python
# get the mean of the price of the each of the house
print('This is the mean value of each of the price of the house within the data:\n', data['price'].mean())
```

```python
# get the median of the data
print('This is the median vale of the data:\n', data.median(), '\n')
```

```python
# get the median price of each of the house
print('This is the median value of each of the price of the house within the data:\n', data['price'].median(),'\n')
```

```python
# get the variance of the data
print('This is the variance of the data:\n', data.var(), '\n')
```

```python
# get the variance of the price
print('This is the variance of the price:\n', data['price'].var(),'\n')
```

### visualization using  histograms, scatter plots, and box plots.

```python
# visualization using hitogram
plt.figure(figsize=(8,5))
sns.histplot(data["price"], bins=30, kde=True)
plt.title("Distribution of House Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

print('Insight:\n')

print(' Data is skewed\n')

print('Detects outliers visually')
```

```python
# scatter plot
plt.figure(figsize=(8,5))
sns.scatterplot(x=data["lotSize"], y=data["price"])
plt.title("price vs lotSize")
plt.xlabel("lotSize")
plt.ylabel("price")
plt.show()


print('Insight:\n')

print('Shows relationship between variables\n')

print('correlation')
```

```python
# boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data=data[["price", "bedRooms", "bathRooms"]])
plt.title("Boxplot of Numerical Features")
plt.show()
```

### Report

- The histogram shows that house prices are right-skewed, indicating a few very expensive properties.

- The scatter plot reveals a positive relationship between lotsize and price.

- The box plot highlights the presence of outliers in the price feature, which may affect model performance.

#### Identify correlations between numerical features using  a correlation matrix.

```python
corr = data.corr(numeric_only=True)
print(corr)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
# most important features that affect price
corr["price"].sort_values(ascending=False)
```

```python
# important_features
important_features = corr["price"].abs().sort_values(ascending=False)
print(important_features)
```

#### 📝 Report You Can Use

- Correlation analysis was performed to understand relationships between numerical features. 
- The results showed that square footage and number of bedrooms have strong positive correlations with house
- price, indicating they are key predictors. 
- Some features exhibited weak correlation and may have limited impact on prediction. 
- Additionally, high -correlation between certain independent variables suggests potential multicollinearity,
- which should be considered during model development.
