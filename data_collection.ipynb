### import required libraries

```python
# import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

### Scrap the target website

```python
# getting requests 
url = "https://www.propertypro.ng/property-for-sale"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, "html.parser")

prices = []
locations = []

for listing in soup.find_all("div", class_="single-room-text"):

    price = listing.find("span", class_="price")
    location = listing.find("span", class_="address")

    if price and location:
        prices.append(price.text.strip())
        locations.append(location.text.strip())

df = pd.DataFrame({
    "Price": prices,
    "Location": locations
})

print(df.head())

df.to_csv("clean_data.csv", index=False)
```
